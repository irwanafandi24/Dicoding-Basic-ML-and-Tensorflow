### Unsupervised - Kmeans

```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
#https://www.kaggle.com/vjchoudhary7/customer-segmentation-tutorial-in-python
df = pd.read_csv('mall_customer.csv')
df.head()
```

```
df = df.rename(columns={'CustomerID':'id','Gender':'gender', 'Age':'age','Annual Income (k$)':'anual','Spending Score (1-100)':'spending'})
df.head()
```

```
df['gender']=df['gender'].map({'Male':0,'Female':1})
df.head()
```

```
from sklearn.cluster import KMeans
X = df.drop(['id'], axis=1)
clusters = []
for i in range(1,11):
  km = KMeans(n_clusters=i).fit(X)
  clusters.append(km.inertia_)
```

```
#creating inertia plot
fig, ax = plt.subplots(figsize=(8,4))
sns.lineplot(x=list(range(1,11)), y=clusters, ax=ax)
plt.xlabel('Cluster')
plt.ylabel('Inertia')
ax.set_title('Find Elbow Number')
plt.show()
```

```
#membuat kmean object
km5 = KMeans(n_clusters=5).fit(X)
#menambahkan label kolom
X['label']=km5.labels_
#membuat plot kmeans dengan 5 cluster
plt.figure(figsize=(8,4))
sns.scatterplot(X['anual'], X['spending'], hue=X['label'],
                palette=sns.color_palette('hls', 5))
plt.title('Kmeans dengan 5 Cluster')
plt.show()
```

### PCA

```
#materi https://www.dicoding.com/academies/184/tutorials/8437?from=8432
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```
iris = datasets.load_iris()
atribut = iris.data
label = iris.target
#split the data
x_train,x_test,y_train,y_test = train_test_split(atribut, label, test_size=0.2)
```

```
from sklearn import tree
```

```
dec_tree = tree.DecisionTreeClassifier()
first_m = dec_tree.fit(x_train, y_train)
first_m.score(x_test,y_test)
```

```
#creating PCA Object
pca = PCA(n_components=4)
#applied PCA on dataset
pc_attributes = pca.fit_transform(x_train)
#look variance in every atribute
pca.explained_variance_ratio_
```

```
pca =  PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.fit_transform(x_test)
```

```
second_model = dec_tree.fit(x_train_pca, y_train)
second_model.score(x_test_pca,y_test)
```

- point yang didapat adalah, model pertama dibuat tanpa reduksi dengan akurasi 0.96.
- model kedua dibuat dengan melakukan reduksi dimensi dengan akurasi 0.86. Hal ini akan lebih efektif jika kita bermain data dengan dimensi yang sangat tinggi. Tujuan utamanya adalah untuk mengurangi waktu eksekusi program sebanyak mungkin.

### Tensorflow

```
import tensorflow as tf
```

```
#get data and split it into data train and data test
mnist = tf.keras.datasets.fashion_mnist
(X_train,y_train),(X_test,y_test) = mnist.load_data()

#picture normalization, pixel resolution in picture between 0-255, we need to normalization the image to 255
#to make NN easier to learn the data  
X_train, X_test = X_train / 255.0, X_test / 255.0

#Neural Network Architecture
model = tf.keras.models.Sequential([
  #Flatten convert matrix 2D of the pricture into 1 dimension 
  tf.keras.layers.Flatten(input_shape = (28,28)), #input_shape: picture's resolution (28x28)
  #Danse can used as hidden layer or output layer
  tf.keras.layers.Dense(512, activation=tf.nn.relu), #parameter unit is the number of perceptron on a layer
  #the number of unit depands on the number of label on dataset
  #if 2 labels data used sigmoid, but if more then 2 labels used softmax
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

#to make the model can learn from the data, we need to compile the model
model.compile(optimizer=tf.optimizers.Adam(), #can use adam for optimizer
              loss='sparse_categorical_crossentropy', #if just 2 label use " binary cross entropy"
              metrics=['accuracy']) #just show the matrix

#epochs is the number of back propagation
model.fit(X_train, y_train, epochs=30)
```
